```python
feature=train[['Inv_Id','Vendor_Code','GlCoded','Inv_Amt','Item_Description']]
feature.head()
```

```python
y=train['Product_Category'].values
y[0:5]
```

```python
X=feature
X.head()
```

```python
import re
a=[]
for s in X['Vendor_Code']:
        match = re.search(r'\d+.?\d*', s)
        if match:
            a.append(int(match.group()))
print(a)
```

```python
df=pd.DataFrame(a,columns=['VendorCoded'])
df.head()
```

```python
X2=pd.concat([feature,df],axis=1)
X2.head()
```

```python
X3=X2[['Inv_Id','VendorCoded','GlCoded','Inv_Amt']]
X3.head()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split( X3, y, test_size=0.3, random_state=0)
print ('Train set:', X_train.shape,  y_train.shape)
print ('Test set:', X_test.shape,  y_test.shape)
from sklearn import metrics
```

```python
from sklearn.ensemble import RandomForestClassifier
forest = RandomForestClassifier(n_estimators=50 ,random_state=42)
forest.fit(X_train,y_train)
```

```python
y_pred=forest.predict(X_test)
```

```python
print("Accuracy:",metrics.accuracy_score(y_test, y_pred))
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn import metrics
```

```python
Ks = 20
mean_acc = np.zeros((Ks-1))
std_acc = np.zeros((Ks-1))
for n in range(1,Ks):
    
    #Train Model and Predict  
    neigh = KNeighborsClassifier(n_neighbors = n).fit(X_train,y_train)
    yhat=neigh.predict(X_test)
    mean_acc[n-1] = metrics.accuracy_score(y_test, yhat)

    
    std_acc[n-1]=np.std(yhat==y_test)/np.sqrt(yhat.shape[0])

mean_acc
```

```python
plt.plot(range(1,Ks),mean_acc,'g')
plt.fill_between(range(1,Ks),mean_acc - 1 * std_acc,mean_acc + 1 * std_acc, alpha=0.10)
plt.legend(('Accuracy ', '+/- 3xstd'))
plt.ylabel('Accuracy ')
plt.xlabel('Number of Neighbors (K)')
plt.tight_layout()
plt.show()
```

```python
print( "The best accuracy was with", mean_acc.max(), "with k=", mean_acc.argmax()+1)
```
